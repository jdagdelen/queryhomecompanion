```python
from pymatgen.ext.matproj import MPRester

mpr = MPRester()
```

## querying different combinations of elements

```python
# any compound containing both O and Cu

mpr.query({'elements':{'$all':['O', 'Cu']}}, ['band_gap', 'pretty_formula'])
```

```python
# compounds containing ONLY (and both) Cu and O

mpr.query({'elements':{'$all':['O', 'Cu']}, 'nelements':2}, ['band_gap', 'pretty_formula'])
```

```python
# a way simpler way to do the exact same thing (#rachelislearningshit)

mpr.query({"elements":["Cu", "O"]}, ["band_gap", "pretty_formula"])
```

```python
# compounds containing EITHER Cu and O

mpr.query({'elements':{'$in':['Cu', 'O']}}, ['band_gap', 'pretty_formula'])
```

```python
# compounds containing EITHER or BOTH Cu and O

mpr.query({'elements':{'$in':['Cu', 'O']}}, ['band_gap', 'pretty_formula'])
```

```python
# compounds containing EITHER Cu or O but not both???

# mpr.query({'elements':{'$in':['Cu', 'O']}}, ['band_gap', 'pretty_formula']) ???
```

```python
# compounds with the stoichiometry AB2C4

mpr.query({"anonymous_formula": {"A": 1.0, "B": 2.0,"C": 4.0}}, ["task_id", "pretty_formula"])
```

```python
# all cubic compounds

mpr.query({"spacegroup.crystal_system": {"$in":["hexagonal", "cubic"]}}, ["task_id", "pretty_formula", "spacegroup"])
```

```python
# all cubic or hexagonal compounds

mpr.query({"spacegroup.crystal_system": {"$in":["hexagonal", "cubic"]}}, ["task_id", "pretty_formula", "spacegroup"])
```

```python
# all NON cubic or hexagonal compounds

mpr.query({"spacegroup.crystal_system": {"$nin":["hexagonal", "cubic"]}}, ["task_id", "pretty_formula", "spacegroup"])
```

## Evaluation selectors

```python
# query whether or not a particular search parameter exists in materials documentation
# for example, whether the materials has "tags"
# note: you can use True or 1 interchangeably as the value

mpr.query({'tags':{'$exists':1}}, ['band_gap', 'pretty_formula'])
```

```python
mpr.query({'tags':{'$exists': True}}, ['band_gap', 'pretty_formula'])
```

```python
mpr.query({'tags':{'$exists': False}}, ['band_gap', 'pretty_formula'])
```

```python
# $type has a VERY SPECIFIC meaning for each number you give it (e.g. 2 is a string): 

mpr.query({'tags':{'$type': 2}}, ['band_gap', 'pretty_formula', 'tags'])
```

```python
# find a specific text string somewhere in a search parameter!!!!! COOL! Didn't know you could do this!

mpr.query({'tags':{'$regex': 'spinel'}}, ['band_gap', 'pretty_formula', 'tags'])
```

```python
# this doesn't work though... regex has to be a string

mpr.query({'tags':{'$regex': {'$in':['spinel','Spinel']}}}, ['band_gap', 'pretty_formula', 'tags'])
```

```python
# there's a way to search any text using $text... I just don't know how

mpr.query({'$text':{'$search': 'spinel'}}, ['band_gap', 'pretty_formula', 'tags'])
```

```python
# also not sure about $elemMatch

mpr.query({"band_gap": {"$elemMatch":{"$gte": 5}}},
          ["task_id", "pretty_formula", "band_gap"])
```

## Logic

```python
# I'm not sure if "$or" works either??? followed this from mongo website

mpr.query({"$or": [{"band_gap": {"$gt":5}}, {"elements": {"$all":["S", "O", "C", "K"]}}]},
          ["task_id", "pretty_formula", "spacegroup"])
```

```python
# $and seems to work

mpr.query({"$and": [{"band_gap": {"$gt":5}}, {"elements": {"$all":["S", "O", "C", "K"]}}]},
          ["task_id", "pretty_formula", "spacegroup"])
```

```python
# but to my knowledge there's no reason to ever use it, because it's just equivalent to this:

mpr.query({"band_gap": {"$gt":5}, "elements": {"$all":["S", "O", "C", "K"]}},
          ["task_id", "pretty_formula", "spacegroup"])
```

```python
# to invert the query and give cases where query is NOT true
# $not works, but there might be more elegant ways to write expression

mpr.query({"band_gap": { "$not": { "$lt": 7 }}},
          ["task_id", "pretty_formula", "band_gap"])
```

```python
# another example of $not, but seems like we don't need this for mp
# (unless you can think of a more clever use case?)

mpr.query({"elements": { "$not": {"$in":["S"]}}},
          ["task_id", "pretty_formula", "band_gap"])
```

```python
# $nor specifies when neither of these conditions holds
# note: takes a longer time to query than other selectors in MP

mpr.query({"$nor": [{"band_gap": {"$lt":5}}, {"elements": {"$all":["S", "O", "C", "K"]}}]},
          ["task_id", "pretty_formula", "band_gap"])
```
